```python
# üß≠ 3) Ïù∏ÌÑ∞ÎûôÌã∞Î∏å ÌëúÎ°ú Î≥¥Í∏∞ (Ïä§ÌÅ¨Î°§/Ï†ïÎ†¨/Í≤ÄÏÉâ Í∞ÄÎä•) ‚ú®
from google.colab import data_table
data_table.enable_dataframe_formatter()

# Ï†ÑÏ≤¥Îäî ÎÑàÎ¨¥ ÌÅ¨Î©¥ Î†åÎçîÎßÅÏù¥ ÎäêÎ¶¥ Ïàò ÏûàÏñ¥Ïöî ‚Üí Ïö∞ÏÑ† ÏÉÅÏúÑ 200Ìñâ ÎØ∏Î¶¨Î≥¥Í∏∞
df.head(200)
```

```python
# üß∞ 4) ÏûêÏ£º Ïì∞Îäî Î∑∞(ÏòàÏãú) ‚Äî ÌïÑÏöîÌï† ÎïåÎßå Ïã§ÌñâÌïòÏÑ∏Ïöî üòä

# (a) ÌïµÏã¨ Î©îÌÉÄÎßå Ï∂îÎ†§ Î≥¥Í∏∞
meta_cols = [
    "subject_id","hadm_id","gender","anchor_age","anchor_year","anchor_year_group",
    "admittime","dischtime","admission_type","insurance","hospital_expire_flag"
]
meta_cols = [c for c in meta_cols if c in df.columns]
df[meta_cols].head(50)

# (b) ÌÖçÏä§Ìä∏ ÏÑπÏÖòÎßå ÌõëÏñ¥Î≥¥Í∏∞
text_cols = [
    "chief_complaint","hpi","pmh","fhx","shx","allergies","physical_exam",
    "assessment","hospital_course","discharge_diagnosis","plan","discharge_instructions","meds_on_discharge"
]
text_cols = [c for c in text_cols if c in df.columns]
df[text_cols].head(10)

# (c) ÌÇ§ÏõåÎìúÎ°ú Í∞ÑÎã® Í≤ÄÏÉâ(Ïòà: 'sepsis'Í∞Ä assessmentÎÇò hospital_courseÏóê Ìè¨Ìï®Îêú Ìñâ)
kw = "sepsis"
mask = False
for col in ["assessment","hospital_course","discharge_diagnosis","plan"]:
    if col in df.columns:
        mask = mask | df[col].fillna("").str.contains(kw, case=False, na=False)
df.loc[mask, ["subject_id","hadm_id"] + [c for c in ["assessment","hospital_course","discharge_diagnosis","plan"] if c in df.columns]].head(50)
```

```python
# (d) JSON ÏöîÏïΩ Ïª¨ÎüºÏùÑ ÌéºÏ≥êÏÑú ÏùºÎ∂Ä ÌÇ§Îßå Î≥¥Í∏∞ (Ïòà: lab_summary_jsonÏóêÏÑú Na/K/Cr)
import json

def pick_labs(js, keys=("Na","K","Cr")):
    try:
        obj = json.loads(js) if isinstance(js, str) else js
        return {k: obj.get(k) for k in keys}
    except Exception:
        return {}

if "lab_summary_json" in df.columns:
    labs_view = df[["subject_id","hadm_id","lab_summary_json"]].copy()
    labs_view["picked"] = labs_view["lab_summary_json"].apply(pick_labs)
    labs_view.head(30)
```
